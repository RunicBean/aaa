## Search Methods
### Solving problems by searching
把状态作为节点，把改变节点的动作作为边，建立搜索路线图。 - PPT P5/6
### Breadth-first search（广度优先）
一级一级向下扩展，每一级将该级别可以出现的所有情况扩展出来之后，再继续下一级。

每次扩展的时候，把新的可能路径加在已有路径后。
### Depth-first search（深度优先）
每次选择一个子节点，不断往深处扩展，知道一条线扩展到无法扩展，再开始下一条线。

> 如果有环存在，则会产生死循环，所以需要先验环。


## Comparisons（时间复杂 / 空间复杂 / 最优化）
在最坏的情况下：
程序执行的时间，O(n) 表示随着数据量增加需要的时间关系，
存储需要的空间，O(n) 表示随着数据量增大需要的空间存储
### Time complexity
BFS / DFS O(b^d)
(b=branch factor, d=depth)

这两个搜索方法的时间复杂度是一样的，都为$O(b^d)$

### Space complexity
BFS
按层级搜索，最坏的情况是把所有点都储存在内存中，所以空间复杂度$O(b^d)$


DFS
按深度搜索，每次搜索完毕一条路线的时候，需要记录每一个节点的所有子节点，也就是b，那么最坏的情况就是到最深的一条线，所以记录的空间容量是$O(bd)$


### Optimality（最优解）
BFS is optimal under certain constraints  
DFS is never optimal

#### BFS optimal conditions
当条件满足任意两点间的costs大于等于0的时候，每次都进行排序(cost最小值)，则可以计算BFS最优解。  
1. 为什么要满足cost大于等于0？  
如果某子路径的cost为负数，则不能确定第一次找到的目标点就是cost最小的，因为可能后续步骤中可能cost更小。



2. 为什么要进行排序？  
要保证每次寻址的条件是cost最小的，最优解才有意义。

## A* search
和BFS类似，区别在于：
1. 进行排序的cost计算方式等于 最初点到当前点的cost 加上 当前点到目标点的estimated cost，也就是估计方式不同。   
2. 每进行一步找寻子节点后，会根据估计的cost来去掉cost较大的路线，也就是剪枝。


estimate cost可以根据直线距离判断，也可以根据具体场景的分析和预测来判断estimated cost。
